```
from pyspark.sql import SparkSession
import sys
spark = SparkSession.builder\
        .master("local")\
        .appName('Pyspark')\
        .getOrCreate()
context = spark.sparkContext
file = context.textFile('dbfs:/FileStore/shared_uploads/sbardar31@gmail.com/word_count-1.txt')
count = file.flatMap(lambda line: line.split(" "))\
        .map(lambda word:(word,1))\
        .reduceByKey(lambda x,y :x+y)
final = count.collect()
for (word,count) in final:
  print(word, ':', count)
```
